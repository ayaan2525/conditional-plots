```python
import pandas as pd
titanic = pd.read_csv('train.csv')
titanic = titanic[["Survived","Pclass","Sex","Age","SibSp","Parch","Fare","Embarked"]]
print("rows before removing null values: %d" %(titanic.shape[0]))
titanic = titanic.dropna()
print("rows before removing null values: %d" %(titanic.shape[0]))
```

```python
titanic.head()
```

```python
import seaborn as sns    # seaborn is commonly imported as `sns`
import matplotlib.pyplot as plt
sns.distplot(titanic["Fare"])
plt.show()
```

```python
#ploting the age
sns.kdeplot(titanic['Age'],shade=True)
plt.xlabel("Age")
plt.show()
```

```python
#Visualizing the differences in age distributions between passengers 
#who survived and those who didn't bycreating a pair of kernel density plots.
g = sns.FacetGrid(titanic,col='Survived', size=6)
g.map(sns.kdeplot,'Age',shade=True)

sns.despine(left=True,bottom=True)
plt.show()
```

```python
#different combinations of unique values in both the Pclass and Survived column
g = sns.FacetGrid(titanic, col="Survived", row="Pclass",size=4)
g = (g.map(sns.kdeplot, "Age", shade=True).set(xlim=(-50,100),ylim=(0,0.04)))
sns.despine(left=True, bottom=True)
plt.show()
```

```python
#different combinations of unique values in Pclass, Sex and Survived column
g = sns.FacetGrid(titanic, col="Survived", row="Pclass",hue='Sex')
g = (g.map(sns.kdeplot, "Age", shade=True).add_legend().set(xlim=(-50, 100), ylim=(0, 0.04)))
sns.despine(left=True, bottom=True)
plt.show()
```
